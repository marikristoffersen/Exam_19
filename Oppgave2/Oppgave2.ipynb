```python
df.unins_clean = df.unins_clean.replace('Insured', 0)
df.unins_clean = df.unins_clean.replace('Uninsured', 1)
```

```python
model = 'unins_clean ~ race_clean + cancerdx_clean + sex_clean + age53x'
propensity = smf.logit(formula=model, data = df).fit()
propensity.summary()
```

```python
df['propensity'] = propensity.predict()
df.groupby('unins_clean')['propensity'].hist(alpha=0.4);
```
